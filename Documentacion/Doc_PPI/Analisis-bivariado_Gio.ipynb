**Relaciones entre Variables:**
- Analizar la relaci√≥n entre pares de variables para identificar patrones o correlaciones.
Utilizar gr√°ficos de dispersi√≥n (scatter plots) y mapas de calor de correlaci√≥n (heatmaps) para evaluar la relaci√≥n entre variables num√©ricas.
- Comparaciones Categ√≥ricas: Explorar c√≥mo las variables categ√≥ricas afectan a las variables num√©ricas, utilizando gr√°ficos de barras y boxplots

- Heatmaps de Correlaci√≥n: permitir√° evaluar la relaci√≥n entre todas las variables de presi√≥n arterial. Esto es √∫til para identificar correlaciones fuertes o d√©biles entre las variables sist√≥licas y diast√≥licas:
  Se muestra una correlaci√≥n positiva fuerte , el valor cerecano a 1 en la celda que cruza sist√≥lica y diast√≥lica indica una correlaci√≥n positiva muy fuerte entre ambas variables. A medida que aumenta la presi√≥n sist√≥lica aumenta la diast√≥lica. y viceverza.


```python
plt.figure(figsize=(5, 5))
sns.heatmap(df_unido[['Systolic','Diastolic', ]].corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n entre variables de presi√≥n arterial')
plt.show()
```

Este Gr√°fico examina la relaci√≥n entre dos variables num√©ricas: la edad y la presi√≥n sist√≥lica. De manera general, en este caso, se agrupan las edades en rangos y se calcula la media de la presi√≥n sist√≥lica en cada grupo de edad, lo que permite visualizar c√≥mo cambia la presi√≥n sist√≥lica en funci√≥n de la edad. La l√≠nea conecta los puntos que representan el promedio de la presi√≥n sist√≥lica en cada grupo de edad, lo permite observar c√≥mo la presi√≥n sist√≥lica var√≠a a medida que las personas envejecen. 
Tendencia: Si la l√≠nea tiene una pendiente ascendente, indica que a medida que la edad aumenta, el promedio de la presi√≥n sist√≥lica tambi√©n aumenta. Si es descendente, significar√≠a que la presi√≥n sist√≥lica disminuye con la edad.
Relaci√≥n: Este gr√°fico muestra de manera visual la relaci√≥n entre la variable edad (agrupada) y la variable presi√≥n sist√≥lica, destacando las tendencias promedias.  

```python
# Filtrar solo las filas correspondientes al sexo femenino ('Female')
df_female = df_limpio[df_limpio['GenderId'] == 'Female']

# Corregir los intervalos de edad y etiquetas para el sexo femenino
df_female['Age_Group'] = pd.cut(df_female['Age'], 
                                bins=[9, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90], 
                                labels=['9-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-60', '61-70', '71-80', '81-90'])

# Agrupar por grupo de edad y calcular el promedio de presi√≥n diast√≥lica para el sexo femenino
age_grouped_female = df_female.groupby('Age_Group')['Diastolic'].mean().reset_index()

# Filtrar solo las filas correspondientes al sexo masculino ('Male')
df_male = df_limpio[df_limpio['GenderId'] == 'Male']

# Corregir los intervalos de edad y etiquetas para el sexo masculino
df_male['Age_Group'] = pd.cut(df_male['Age'], 
                              bins=[9, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90], 
                              labels=['9-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-60', '61-70', '71-80', '81-90'])

# Agrupar por grupo de edad y calcular el promedio de presi√≥n diast√≥lica para el sexo masculino
age_grouped_male = df_male.groupby('Age_Group')['Diastolic'].mean().reset_index()

# Crear gr√°fico con l√≠neas superpuestas
plt.figure(figsize=(12, 6))

# L√≠nea para el sexo femenino
sns.lineplot(x='Age_Group', y='Diastolic', data=age_grouped_female, marker='o', label='Mujeres', color='blue')

# L√≠nea para el sexo masculino
sns.lineplot(x='Age_Group', y='Diastolic', data=age_grouped_male, marker='o', label='Hombres', color='red')

# A√±adir t√≠tulo y etiquetas
plt.title('Promedio de Presi√≥n Diast√≥lica por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Promedio de Presi√≥n Diast√≥lica')

# Mostrar la leyenda
plt.legend()

# Ajustar el dise√±o
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Suponiendo que 'df_limpio' es tu DataFrame con la columna 'Date' en formato (d√≠a/mes/a√±o)
# Primero, convierte la columna 'Date' a tipo datetime
df_limpio['Date'] = pd.to_datetime(df_limpio['Date'], format='%d/%m/%Y')

# Filtrar datos para el a√±o 2024
df_2024 = df_limpio[df_limpio['Date'].dt.year == 2024]

# Extraer el mes de la fecha
df_2024['Month'] = df_2024['Date'].dt.month

# Filtrar datos por g√©nero
df_female = df_2024[df_2024['GenderId'] == 'Female']
df_male = df_2024[df_2024['GenderId'] == 'Male']

# Agrupar por mes y calcular el promedio de presi√≥n diast√≥lica
heatmap_data_female = df_female.groupby('Month')['Diastolic'].mean().reindex(range(1, 13))
heatmap_data_male = df_male.groupby('Month')['Diastolic'].mean().reindex(range(1, 13))

# Crear un DataFrame para el heatmap
heatmap_data = pd.DataFrame({
    'Female': heatmap_data_female,
    'Male': heatmap_data_male
})

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data.T, 
    cmap='coolwarm', 
    annot=True, 
    fmt='.1f', 
    linewidths=.5, 
    cbar_kws={'label': 'Presi√≥n Diast√≥lica Promedio'}
)
plt.title('Promedio de Presi√≥n Diast√≥lica por Mes en 2024 (Hombres y Mujeres)')
plt.xlabel('Mes')
plt.ylabel('G√©nero')
plt.xticks(ticks=range(12), labels=['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.yticks(rotation=0)
plt.show()
```

```python
# Aseg√∫rate de que la columna 'Date' sea del tipo datetime
df_limpio['BirthDate'] = pd.to_datetime(df_limpio['BirthDate'], format='%d/%m/%Y')

# Extraer el a√±o de la fecha
df_limpio['Year'] = df_limpio['BirthDate'].dt.year

# Filtrar y preparar los datos para hombres y mujeres
df_male = df_limpio[df_limpio['GenderId'] == 'Male']
df_female = df_limpio[df_limpio['GenderId'] == 'Female']

# Calcular el promedio de la presi√≥n sist√≥lica por a√±o
avg_systolic_male = df_male.groupby('Year')['Systolic'].mean()
avg_systolic_female = df_female.groupby('Year')['Systolic'].mean()

# Crear DataFrames a partir de las Series
heatmap_data_male = avg_systolic_male.reset_index()
heatmap_data_female = avg_systolic_female.reset_index()

# A√±adir una columna de g√©nero para diferenciaci√≥n en el heatmap
heatmap_data_male['Gender'] = 'Male'
heatmap_data_female['Gender'] = 'Female'

# Concatenar los DataFrames para facilitar la creaci√≥n del heatmap
heatmap_data = pd.concat([heatmap_data_male, heatmap_data_female])

# Filtrar a√±os con datos nulos
valid_years = heatmap_data.groupby('Year')['Systolic'].mean().dropna().index
heatmap_data_filtered = heatmap_data[heatmap_data['Year'].isin(valid_years)]

# Pivotar los datos para la visualizaci√≥n
heatmap_pivot = heatmap_data_filtered.pivot(index='Gender', columns='Year', values='Systolic')

# Verificar los datos del heatmap pivotado
print(heatmap_pivot)

# Crear el heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(
    heatmap_pivot, 
    cmap='coolwarm', 
    annot=True, 
    fmt='.1f', 
    linewidths=.5, 
    annot_kws={"size": 10, "weight": 'bold', "ha": 'center', "va": 'center'},
    cbar_kws={'label': 'Presi√≥n Sist√≥lica Promedio'}
)
plt.title('Presi√≥n Sist√≥lica Promedio por A√±o y G√©nero (Datos V√°lidos)')
plt.xlabel('A√±o')
plt.ylabel('G√©nero')
plt.show()
```

Boxplot: Muestra la distribuci√≥n, como var√≠a la presi√≥n sist√≥lica en diferentes grupos de edad. Los datos se agrupan en rangos de edad (0-20, 21-40, 41-60, etc.), y para cada grupo, el boxplot muestra:

La mediana (l√≠nea dentro de la caja).
El rango intercuart√≠lico (Q1 a Q3, la caja).
Valores at√≠picos (puntos fuera de los "bigotes").
Los "bigotes" representan la dispersi√≥n de los datos sin considerar los outliers.
Puedes ver si la presi√≥n tiende a aumentar en grupos de mayor edad, o si hay mucha variabilidad dentro de un grupo.

```python
df_limpio['Age_Group'] = pd.cut(df_limpio['Age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])

plt.figure(figsize=(14, 7))
sns.boxplot(x='Age_Group', y='Systolic', data=df_limpio)
plt.title('Distribuci√≥n de la Presi√≥n Sist√≥lica por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Systolic')
plt.show()
```
