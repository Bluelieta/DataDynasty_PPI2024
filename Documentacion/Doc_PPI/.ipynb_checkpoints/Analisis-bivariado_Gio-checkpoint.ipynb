Este Gráfico examina la relación entre dos variables numéricas: la edad y la presión sistólica. De manera general, en este caso, se agrupan las edades en rangos y se calcula la media de la presión sistólica en cada grupo de edad, lo que permite visualizar cómo cambia la presión sistólica en función de la edad. La línea conecta los puntos que representan el promedio de la presión sistólica en cada grupo de edad, lo permite observar cómo la presión sistólica varía a medida que las personas envejecen. 
Tendencia: Si la línea tiene una pendiente ascendente, indica que a medida que la edad aumenta, el promedio de la presión sistólica también aumenta. Si es descendente, significaría que la presión sistólica disminuye con la edad.
Relación: Este gráfico muestra de manera visual la relación entre la variable edad (agrupada) y la variable presión sistólica, destacando las tendencias promedias.  

```python
# Filtrar solo las filas correspondientes al sexo femenino ('Female')
df_female = df_limpio[df_limpio['GenderId'] == 'Female']

# Corregir los intervalos de edad y etiquetas para el sexo femenino
df_female['Age_Group'] = pd.cut(df_female['Age'], 
                                bins=[9, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90], 
                                labels=['9-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-60', '61-70', '71-80', '81-90'])

# Agrupar por grupo de edad y calcular el promedio de presión diastólica para el sexo femenino
age_grouped_female = df_female.groupby('Age_Group')['Diastolic'].mean().reset_index()

# Filtrar solo las filas correspondientes al sexo masculino ('Male')
df_male = df_limpio[df_limpio['GenderId'] == 'Male']

# Corregir los intervalos de edad y etiquetas para el sexo masculino
df_male['Age_Group'] = pd.cut(df_male['Age'], 
                              bins=[9, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90], 
                              labels=['9-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-60', '61-70', '71-80', '81-90'])

# Agrupar por grupo de edad y calcular el promedio de presión diastólica para el sexo masculino
age_grouped_male = df_male.groupby('Age_Group')['Diastolic'].mean().reset_index()

# Crear gráfico con líneas superpuestas
plt.figure(figsize=(12, 6))

# Línea para el sexo femenino
sns.lineplot(x='Age_Group', y='Diastolic', data=age_grouped_female, marker='o', label='Mujeres', color='blue')

# Línea para el sexo masculino
sns.lineplot(x='Age_Group', y='Diastolic', data=age_grouped_male, marker='o', label='Hombres', color='red')

# Añadir título y etiquetas
plt.title('Promedio de Presión Diastólica por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Promedio de Presión Diastólica')

# Mostrar la leyenda
plt.legend()

# Ajustar el diseño
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Suponiendo que 'df_limpio' es tu DataFrame con la columna 'Date' en formato (día/mes/año)
# Primero, convierte la columna 'Date' a tipo datetime
df_limpio['Date'] = pd.to_datetime(df_limpio['Date'], format='%d/%m/%Y')

# Filtrar datos para el año 2024
df_2024 = df_limpio[df_limpio['Date'].dt.year == 2024]

# Extraer el mes de la fecha
df_2024['Month'] = df_2024['Date'].dt.month

# Filtrar datos por género
df_female = df_2024[df_2024['GenderId'] == 'Female']
df_male = df_2024[df_2024['GenderId'] == 'Male']

# Agrupar por mes y calcular el promedio de presión diastólica
heatmap_data_female = df_female.groupby('Month')['Diastolic'].mean().reindex(range(1, 13))
heatmap_data_male = df_male.groupby('Month')['Diastolic'].mean().reindex(range(1, 13))

# Crear un DataFrame para el heatmap
heatmap_data = pd.DataFrame({
    'Female': heatmap_data_female,
    'Male': heatmap_data_male
})

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data.T, 
    cmap='coolwarm', 
    annot=True, 
    fmt='.1f', 
    linewidths=.5, 
    cbar_kws={'label': 'Presión Diastólica Promedio'}
)
plt.title('Promedio de Presión Diastólica por Mes en 2024 (Hombres y Mujeres)')
plt.xlabel('Mes')
plt.ylabel('Género')
plt.xticks(ticks=range(12), labels=['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.yticks(rotation=0)
plt.show()
```

```python
# Asegúrate de que la columna 'Date' sea del tipo datetime
df_limpio['BirthDate'] = pd.to_datetime(df_limpio['BirthDate'], format='%d/%m/%Y')

# Extraer el año de la fecha
df_limpio['Year'] = df_limpio['BirthDate'].dt.year

# Filtrar y preparar los datos para hombres y mujeres
df_male = df_limpio[df_limpio['GenderId'] == 'Male']
df_female = df_limpio[df_limpio['GenderId'] == 'Female']

# Calcular el promedio de la presión sistólica por año
avg_systolic_male = df_male.groupby('Year')['Systolic'].mean()
avg_systolic_female = df_female.groupby('Year')['Systolic'].mean()

# Crear DataFrames a partir de las Series
heatmap_data_male = avg_systolic_male.reset_index()
heatmap_data_female = avg_systolic_female.reset_index()

# Añadir una columna de género para diferenciación en el heatmap
heatmap_data_male['Gender'] = 'Male'
heatmap_data_female['Gender'] = 'Female'

# Concatenar los DataFrames para facilitar la creación del heatmap
heatmap_data = pd.concat([heatmap_data_male, heatmap_data_female])

# Filtrar años con datos nulos
valid_years = heatmap_data.groupby('Year')['Systolic'].mean().dropna().index
heatmap_data_filtered = heatmap_data[heatmap_data['Year'].isin(valid_years)]

# Pivotar los datos para la visualización
heatmap_pivot = heatmap_data_filtered.pivot(index='Gender', columns='Year', values='Systolic')

# Verificar los datos del heatmap pivotado
print(heatmap_pivot)

# Crear el heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(
    heatmap_pivot, 
    cmap='coolwarm', 
    annot=True, 
    fmt='.1f', 
    linewidths=.5, 
    annot_kws={"size": 10, "weight": 'bold', "ha": 'center', "va": 'center'},
    cbar_kws={'label': 'Presión Sistólica Promedio'}
)
plt.title('Presión Sistólica Promedio por Año y Género (Datos Válidos)')
plt.xlabel('Año')
plt.ylabel('Género')
plt.show()
```

Implicaciones: Esta fuerte correlación sugiere que ambas medidas de presión arterial están estrechamente relacionadas y probablemente reflejan patrones similares de comportamiento en los datos.
```python
mayores_60_df = df_limpio[df_limpio['Age'] > 80]

sns.histplot(mayores_60_df['Systolic'], bins=20, kde=True)
plt.title('Distribución de la Presión Sistólica en Personas Mayores de 80 Años')
plt.xlabel('Presión Sistólica')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar los datos para mostrar solo aquellos con edad mayor a 60 años
mayores_30_df = df_limpio[df_limpio['Age'] > 30]

# Crear el gráfico de líneas
sns.lineplot(x='Age', y='Systolic', data=mayores_30_df)
plt.title('Presión Sistólica para Personas Mayores de 30 Años')
plt.show()
```
